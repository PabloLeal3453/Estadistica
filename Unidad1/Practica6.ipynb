```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
```

```python
auto = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

```python
auto.head()
```

```python
auto.info()
```

#1.- Estadsticas descriptivas bsicas y medidas de tendencia central

```python
#Precio minimo
auto.price.min()
```

```python
#Precio maximo
auto.price.max()
```

```python
#esto lo hiso la ia, porque en media habia un error que daba nan.
auto['price'] = pd.to_numeric(auto['price'], errors='coerce')
auto = auto.dropna(subset=['price'])
#---------------------------------------------------------------#
moda=mode(auto.price)
media=mean(auto.price)
mediana=median(auto.price)
desviacion=stdev(auto.price)
print("Moda: ", moda)
print("Media: ", media)
print("Mediana: ", mediana)
print("Desviacion: ", desviacion)
```

```python
#Determinar los estadsticos y cuartiles con la funcin describe()
auto.price.describe()
```

#2.-Histograma de precios de los automviles

```python
#Realizar histograma de precios de los automoviles
pyplot.hist(auto.price, bins=9)
pyplot.xlabel("Precio")
pyplot.ylabel("Frecuencia")
pyplot.title("Histograma de precios de los automviles")
pyplot.show()
```

```python
# Conclusiones, en conclusion entiendo que se fabrican menos carros con rango de
# precio mayor, yo creo para que se venda mas, ya que hay muy poca gente con
# tanto estado economico superior como para comprar autos caros
```

#3.-diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo de vehculo (body style)

```python
sns.boxplot(x='body-style', y='price', data=auto)
pyplot.xlabel("Tipo de vehiculo")
pyplot.ylabel("Precio")
pyplot.title("Diagrama de caja y bigotes ")
pyplot.show()
```

```python
# Esto nos puede ayudar para saber cuales pueden ser las opciones, entre tipo de
# vehiculos mas baratos, o mas caros, segun convenga.
```

#4.-Diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo traccin ruedas (drive wheels)

```python
sns.boxplot(x='drive-wheels', y='price', data=auto)
pyplot.xlabel("Tipo de traccion de ruedas")
pyplot.ylabel("Precio")
pyplot.title("Diagrama de caja y bigotes ")
pyplot.show()
```

```python
# Esto en cambio nos dice acerca de los tipo de tracciones de ruedas y sus
# diferentes rangos de precios, y vemos que el precio que mas se vende es entre
# 5000 y 25000
```

#5.-Diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y la ubicacin del motor (engine-location)

```python
sns.boxplot(x='engine-location', y='price', data=auto)
pyplot.xlabel("Ubicacion del motor")
pyplot.ylabel("Precio")
pyplot.title("Diagrama de caja y bigotes ")
pyplot.show()
```

```python
# Con esto entiendo que la manera en que se interpretan los datos, nos sirven
# para verificar precios de los motores, y ademas saber su ubicacion, para
# propositos desconocidos, pero el diagrama de caja y bigotes,nos ayuda a saber
# mas rapido esta informacion, o visualizarla.
```
