```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('grupo1.csv')
df1.head()

plt.figure(figsize=(6,4))
sns.histplot(df1['Calificacion'], bins=10, kde=True, color='skyblue')
plt.title('Histograma de Calificaciones grupo 1')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

X1 = df1[['HorasEstudio']]
y1 = df1['Calificacion']
modelo1 = LinearRegression()
modelo1.fit(X1, y1)
pendiente1 = modelo1.coef_[0]
intercepto1 = modelo1.intercept_
print(f"Modelo1: y = {intercepto1:.2f} + {pendiente1:.2f}x")
r, _ = pearsonr(df1['HorasEstudio'], df1['Calificacion'])
print(f"Coeficiente de correlación de Pearson: {r:.2f}")
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df1)
plt.plot(df1['HorasEstudio'], modelo1.predict(X1), color='red', label='Línea de regresión')
plt.title('Relación entre Horas de Estudio y Calificación Final grupo 1')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.show()
```

##1. ¿Cuál es la interpretación del coeficiente de regresión en relación con el tiempo de estudio y las calificaciones finales?
La pendiente va hacia abajo lo cual se me hace raro porque lo lógico sería que mientras más estudien mejor les vaya pero en este caso no parece haber mucho cambio en las calificaciones.

##2. ¿Cómo se interpreta el coeficiente de correlación en cuanto a la fuerza y dirección de la relación entre las variables?
Está cerca de cero así que no hay una relación clara entre las horas de estudio y las calificaciones.

##3. ¿El modelo de regresión lineal es estadísticamente significativo?
No lo es porque no se ve una tendencia clara ni nada que diga que hay una relación fuerte entre las variables.

##4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales en función del tiempo de estudio?
No es muy confiable porque los datos están muy dispersos y seguro hay otros factores que influyen más que solo estudiar.
