```python
from statistics import *
#Lista de calificaciones del grupo a de probabilidad y estadistica.

grupoa = [70, 70, 65,90, 100, 95,45 ,80, 73, 20, 100, 96]
print (grupoa)

#La funcion type nos devuelve el tipo de dato de la variable.
type(grupoa)

#Sacar la media  de las calificaciones con una formula
promedio = sum(grupoa)/len(grupoa)
print("promedio 1:",promedio)

#Utilizar la funcion mean que esta en la libreria statistics para sacar
#el promedio

promedio2 = mean(grupoa)
print("promedio con libreria:",promedio2)

#Determinar el valor de la mediana con la funcion mediana.
mediana = median(grupoa)
print("mediana:",mediana)

#Determinar el valor de la moda con la funcion mode.
moda = mode(grupoa)
print("moda:",moda)

#En caso de que exista mas de una moda, se puede emplear la funcion multimoda

multimoda = multimode(grupoa)
print("multimoda:",multimoda)

```

```python
#Definimos una variable llamada tiempos100mts de tipo lista, que contiene los
#tiempos que tardaron los 10 alumnos para correr 100 mts
tiempos100mts = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.10, 17.20]

#Desplegamos los valores de la lista
print("Tiempo de cada alumnos:",tiempos100mts)

#Usamos la funcion mean para determinar el tiempo promedio de la lista de valores

print("Promedio de tiempos:", mean(tiempos100mts))
#Usamos la funcion median para determinar el tiempo promedio de la lista de valores
#Definimos una variable llamada mediana para calucular la mediana de la lista con la funcion
#median

mediana = median(tiempos100mts)
print("Mediana de tiempos:", mediana)

#definimos una variable llamada moda para calcular la moda de la lista con la funcion
#mode

moda = mode(tiempos100mts)
print("Moda de tiempos:", moda)

multimoda = multimode(tiempos100mts)
print("Multimoda de tiempos:", multimoda)

```

```python
#Histogramas de 2 grupos
#libreria de python para graficar valores
import random
fabrica1=[]
for i in range(100):
  numero_aleatorio= round( random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)

fabrica2=[]
for i in range(100):
  numero_aleatorio= round( random.uniform(10,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
# Libreria de python para graficar valores
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.hist("Produccion diaria de la fabrica 1:")
plt.show()
plt.hist(fabrica2)
plt.hist("Produccion diaria de la fabrica 2:")
plt.show()
media1= mean(fabrica1)
mediana1= median(fabrica1)
moda1= mode(fabrica1)
print("")
print ("Valores de la fabrica 1:")
print("")
print("Media fabrica 1:", media1)
print("Mediana fabrica 1:", mediana1)
print("Moda fabrica 1:", moda1)
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("")
print("Valores de la fabrica 2:")
print("")
print("Media fabrica 2:", media2)
print("Mediana fabrica 2:", mediana2)
print("Moda fabrica 2:", moda2)
#GENERAR 2 SUBPLOT DE  1 un renglon por dos columnas para desplegar los histogramas
#una al lado de la otra.
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
plt.hist(fabrica1, bins=10, alpha=0.2, color='b',edgecolor="black")
plt.axvline(media1, color = 'b', linestyle = 'dashed', linewidth = 2, label= 'Media fabrica 1')
plt.axvline(mediana1, color = 'r', linestyle = 'dashed', linewidth = 2, label= 'Mediana fabrica 1')
plt.axvline(moda1, color = 'g', linestyle = 'dashed', linewidth = 2, label= 'Moda fabrica 1')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Produccion diaria de la fabrica 1')
plt.legend()
plt.show()
plt.hist(fabrica2, bins=10, alpha=0.2, color='b',edgecolor="black")
plt.axvline(media2, color = 'b', linestyle = 'dashed', linewidth = 2, label= 'Media fabrica 2')
plt.axvline(mediana2, color = 'r', linestyle = 'dashed', linewidth = 2, label= 'Mediana fabrica 2')
plt.axvline(moda2, color = 'g', linestyle = 'dashed', linewidth = 2, label= 'Moda fabrica 2')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Produccion diaria de la fabrica 2')
plt.legend()
plt.show()


```
